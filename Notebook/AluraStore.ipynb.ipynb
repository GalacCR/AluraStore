### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import warnings

# Ignorar todos los warnings
warnings.filterwarnings("ignore")
```

```python
# === EXPLORACIÓN INICIAL DE LOS 4 DATASETS ===

# Lista de los dataframes
datasets = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for name, df in datasets.items():
    print(f"\n===== {name} =====")
    print("Shape:", df.shape)          # filas x columnas
    print("\nPrimeras filas:")
    display(df.head(3))                # muestra primeras filas

    print("\nInformación general:")
    display(df.info())                  # tipos de datos y nulos

    print("\nPorcentaje de valores nulos por columna:")
    display((df.isna().mean() * 100).round(2).sort_values(ascending=False))
```

#1. Análisis de facturación



```python
# === INGRESO TOTAL POR TIENDA ===

col_precio = "Precio"

ingresos = {
    "Tienda 1": tienda[col_precio].sum(),
    "Tienda 2": tienda2[col_precio].sum(),
    "Tienda 3": tienda3[col_precio].sum(),
    "Tienda 4": tienda4[col_precio].sum()
}

# Convertir a DataFrame
df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso_Total"])

display(df_ingresos)
```

# 2. Ventas por categoría

```python
# === CANTIDAD DE PRODUCTOS VENDIDOS POR CATEGORÍA EN CADA TIENDA ===

cantidad_productos_vendidos = {}

for name, df in datasets.items():
    print(f"\n===== {name} =====")

    # Agrupar por categoría y contar número de ventas
    ventas_cat = df.groupby("Categoría del Producto").size().reset_index(name="cantidad")

    # Ordenar de mayor a menor
    ventas_cat = ventas_cat.sort_values(by="cantidad", ascending=False)

    # Guardar resultado
    cantidad_productos_vendidos[name] = ventas_cat

    # Mostrar las categorías más vendidas
    display(ventas_cat.head(8))

```

# 3. Calificación promedio de la tienda


```python
# === CALIFICACIÓN PROMEDIO DE LOS CLIENTES POR TIENDA ===

calificaciones_promedio = {}

for name, df in datasets.items():
    # Calificación promedio
    promedio = df["Calificación"].mean()
    calificaciones_promedio[name] = promedio

# Convertir a DataFrame
df_calificaciones = pd.DataFrame(list(calificaciones_promedio.items()), columns=["Tienda", "Calificación_Promedio"])
display(df_calificaciones)
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# === PRODUCTOS MÁS Y MENOS VENDIDOS POR TIENDA ===

for name, df in datasets.items():
    print(f"\n===== {name} =====")

    # Contar ventas por producto
    ventas_producto = df.groupby("Producto").size().reset_index(name="cantidad")

    # Productos más vendidos (top 10)
    top_10 = ventas_producto.sort_values(by="cantidad", ascending=False).head(10)
    print("Top 10 productos más vendidos:")
    display(top_10)

    # Productos menos vendidos (top 10)
    bottom_10 = ventas_producto.sort_values(by="cantidad", ascending=True).head(10)
    print("Top 10 productos menos vendidos:")
    display(bottom_10)
```

# 5. Envío promedio por tienda

```python
# === COSTO DE ENVÍO PROMEDIO POR TIENDA ===

costo_envio_promedio = {}

for name, df in datasets.items():
    promedio = df["Costo de envío"].mean()
    costo_envio_promedio[name] = promedio

# Convertir a DataFrame
df_costo_envio = pd.DataFrame(list(costo_envio_promedio.items()), columns=["Tienda", "Costo_Envio_Promedio"])
display(df_costo_envio)
```

# Visualización

```python
# 1) Función para graficar productos más y menos vendidos
def graficar_productos_ventas(df, tienda, top_n=10):
    # Contar ventas por producto
    ventas_producto = df.groupby("Producto").size().reset_index(name="cantidad")

    # Productos más y menos vendidos
    top_10 = ventas_producto.sort_values(by="cantidad", ascending=False).head(top_n)
    bottom_10 = ventas_producto.sort_values(by="cantidad", ascending=True).head(top_n)

    # Gráfico Top 10 productos más vendidos
    plt.figure(figsize=(10,6))
    sns.barplot(x="cantidad", y="Producto", data=top_10, palette="viridis")
    plt.title(f"Top {top_n} Productos Más Vendidos - {tienda}")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()

    # Gráfico Top 10 productos menos vendidos
    plt.figure(figsize=(10,6))
    sns.barplot(x="cantidad", y="Producto", data=bottom_10, palette="magma")
    plt.title(f"Top {top_n} Productos Menos Vendidos - {tienda}")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()

# Aplicar la función a las 4 tiendas
for name, df in datasets.items():
    graficar_productos_ventas(df, name)
```

```python
# ----------------------------
# 2) Gráfico de barras: Ingresos por tienda
# ----------------------------
plt.figure(figsize=(8,5))
sns.barplot(x="Tienda", y="Ingreso_Total", data=df_ingresos, palette="Blues_d")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingreso Total")
plt.xlabel("Tienda")
plt.show()
```

```python
# ----------------------------
# 3) Gráfico de dispersión: Calificación promedio vs. Costo de envío promedio
# ----------------------------
# Unir calificación y costo de envío por tienda
df_satisfaccion_envio = df_calificaciones.merge(df_costo_envio, on="Tienda")

plt.figure(figsize=(8,5))
sns.scatterplot(x="Costo_Envio_Promedio", y="Calificación_Promedio", data=df_satisfaccion_envio, hue="Tienda", s=200)
plt.title("Relación entre Costo de Envío Promedio y Calificación Promedio")
plt.xlabel("Costo de Envío Promedio")
plt.ylabel("Calificación Promedio")
plt.legend()
plt.show()
```

# Informe de Identificación de la Tienda Menos Eficiente

## Introducción

El objetivo de este análisis es determinar cuál de las cuatro tiendas representa la menor eficiencia, para que el Sr. Juan pueda decidir cual tienda le beneficia más vender.  
Se evaluaron factores clave: ingresos, categorías de productos, productos más vendidos, calificación de clientes y costos de envío.

## Ingresos Totales por Tienda

**Observación:** Tienda 4 tiene el ingreso más bajo, aproximadamente un 10% menos que la Tienda 1.

**Interpretación:** Menor flujo de ventas → menor eficiencia financiera.

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Tienda", y="Ingreso_Total", data=df_ingresos, palette="Reds_d")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingreso Total")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```

## Categorías de Productos Más Vendidas

**Observación:** Tienda 4 tiene menor volumen en categorías de alta demanda como Muebles y Electrónicos, y una proporción mayor de categorías menos populares.

**Interpretación:** Portafolio menos alineado con las preferencias de los clientes → menor rotación y eficiencia.


```python
for name, df in cantidad_productos_vendidos.items():
    plt.figure(figsize=(10,6))
    sns.barplot(x="cantidad", y="Categoría del Producto", data=df.head(8), palette="magma")
    plt.title(f"Top Categorías Más Vendidas - {name}")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Categoría del Producto")
    plt.tight_layout()
    plt.show()
```

## Calificación Promedio de Clientes por Tienda

**Observación:** Tienda 4 tiene buena calificación promedio (4.00), pero ligeramente inferior a las tiendas mejor evaluadas.

**Interpretación:** Aunque la satisfacción es adecuada, no compensa su menor volumen de ventas.


```python
plt.figure(figsize=(8,5))
sns.barplot(x="Tienda", y="Calificación_Promedio", data=df_calificaciones, palette="Oranges")
plt.title("Calificación Promedio de Clientes por Tienda")
plt.ylabel("Calificación Promedio")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```

## Productos Más y Menos Vendidos por Tienda

**Observación:** Tienda 4 tiene menos productos dentro del Top 10 de ventas y más productos en los 10 menos vendidos.

**Interpretación:** Menor diversidad de productos exitosos → riesgo de ventas concentradas y menor atractivo para compradores.


```python
for name, df in datasets.items():
    ventas_producto = df.groupby("Producto").size().reset_index(name="cantidad")
    top_10 = ventas_producto.sort_values(by="cantidad", ascending=False).head(10)
    bottom_10 = ventas_producto.sort_values(by="cantidad", ascending=True).head(10)

    # Gráfico productos mas vendidos
    plt.figure(figsize=(10,6))
    sns.barplot(x="cantidad", y="Producto", data=top_10, palette="viridis")
    plt.title(f"Top 10 Productos Más Vendidos - {name}")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()

    # Gráfico productos menos vendidos
    plt.figure(figsize=(10,6))
    sns.barplot(x="cantidad", y="Producto", data=bottom_10, palette="magma")
    plt.title(f"Top 10 Productos Menos Vendidos - {name}")
    plt.xlabel("Cantidad de ventas")
    plt.ylabel("Producto")
    plt.tight_layout()
    plt.show()
```

## Coste de Envío Promedio por Tienda

**Observación:** Los costos de envío de Tienda 4 son comparables con otras tiendas.

**Interpretación:** No representa ventaja competitiva, y no compensa las debilidades en ingresos y ventas.


```python
plt.figure(figsize=(8,5))
sns.barplot(x="Tienda", y="Costo_Envio_Promedio", data=df_costo_envio, palette="coolwarm")
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```

## Conclusión y Recomendación

Tras analizar los datos:

1. **Ingresos:** Tienda 4 tiene el menor ingreso total, lo que indica menor eficiencia financiera.  
2. **Ventas por categoría y producto:** Tienda 4 muestra menor volumen en categorías clave y más productos de los menos vendidos, evidenciando baja rotación.  
3. **Satisfacción del cliente:** Adecuada (4.00), pero no suficiente para compensar su menor desempeño.  
4. **Costos de envío:** Comparables con otras tiendas, sin ventaja significativa.  

**Recomendación:**  
Se sugiere considerar **Tienda 4 como la menos eficiente**, candidata a **vender**, permitiendo liberar capital y enfocar recursos en tiendas más rentables o en nuevos negocios.  

Esta decisión se respalda en los ingresos, la rotación de productos y la alineación del portafolio con la demanda de los clientes.
